```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import re
```

```python
def inclusive_index(lst, purpose): # 部分一致のindexを返す
    for i, e in enumerate(lst):
        if e.startswith(purpose): 
          return i
    
    return False
```

```python
title_links = []
url_classes = ["ars_al", "ars_fl", "ars_cc", "ars_cs", "ars_hs", "ars_lc", "ars_ic", "ars_ms", 
              # "hcul", "hsoc", "hetc", "edu_kokugo", "edu_shakai", "edu_sugaku", "edu_info", "edu_rika", 
             # "edu_ongaku", "edu_bijutsu", "edu_hoken", "edu_gijutsu", "edu_kasei", "edu_eigo", 
              #"edu_gakko", "edu_tokushu", "edu_yoji", "edu_joho", "edu_shougai", "edu_hattatsu", "edu_kyoushoku", 
              # "edu_kyokyo", "edu_seikatsu", "edu_jicchi", "edu_sonota", 
              # "med", "nur", "eng_com", "eng_ovall", "mech", "elec", "chem", "info", "arch", "phys", "eng_free", 
              # "biob", "hed"
               ]
for i in range(len(url_classes)):
  url = "https://syllabus.mie-u.ac.jp/?action=list&class=" + url_classes[i] + "#sresult"
  r = requests.get(url)
  time.sleep(1)
  soup = BeautifulSoup(r.text, "html.parser")
  get_a = soup.find_all("a", target="_blank")
  #get_a.pop(0)

  for i in range(len(get_a)-1):
      try:
          link_ = get_a[i].get("href")
          title_links.append(link_)
      except:
          pass
```

```python
strs = ["授業科目名", "授業テーマ", "選択・必修", "開放科目", "開講学期", "開講時間", "授業の方法", "成績評価方法と基準", "授業の特徴", "授業改善の工夫" , "受講要件", "開講区分", "分野", "担当教員", "授業の概要", "学修の目的", "学修の到達目標", 
    "教科書", "参考書", "予め履修が望ましい科目", "発展科目", "オフィスアワー", "その他", "キーワード", "Key Word(s)", "学修内容"]
nums = list(range(len(strs)))
temp = list(range(len(title_links)))
df = pd.Series()
CLS = list(range(len(strs)))
```

```python
for i in range(len(title_links)):
    title_link = title_links[i]
    print(str(i+1)+"科目目",title_link)
    r = requests.get("https://syllabus.mie-u.ac.jp/" + title_link)
    time.sleep(1)
    soup = BeautifulSoup(r.text, "html.parser")
    clses = []
    cls = soup.find_all('tr')
    for j in range(len(cls)):
      cls[j] = cls[j].text
      cls[j] = cls[j].split('\n')
      clses.append(''.join(cls[j]))

    for t in range(len(nums)):
      nums[t] = False
    for m in range(len(strs)):
      CLS[m] = ''
      nums[m] = inclusive_index(clses, strs[m])
      if nums[m]:
        clses[nums[m]] = clses[nums[m]].replace('\u3000', '')
        clses[nums[m]] = clses[nums[m]].replace(strs[m], '')
        CLS[m] = clses[nums[m]]
    temp[i] = pd.Series(CLS, name=str(i+1), index=strs)
    print(temp[i][0])
    print(CLS)
```

```python
df = pd.Series()
for i in range(len(temp)):
  if i == 0:
    df = temp[0]
  else:
    df = pd.concat([df, temp[i]], axis = 1)
df = pd.DataFrame(df, index=strs)
df = df.transpose()
#df.to_csv("scraping_result.csv", encoding='utf_8_sig') 
df.to_csv("full_scraping_result.csv", encoding='utf_8_sig') 
```
