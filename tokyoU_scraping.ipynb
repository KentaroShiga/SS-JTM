```python
def inclusive_index(lst, purpose): # 部分一致のindexを返す
    for i, e in enumerate(lst):
        if e.startswith(purpose): 
          return i
    
    return False
```

```python
i = 0
title_links = []
while True:
  i += 1
  url = f"https://catalog.he.u-tokyo.ac.jp/result?q=&type=jd&faculty_id=&facet=%7B%22course_language_codes%22%3A%5B%2201%22%5D%7D&page={i}"
  print(f"{i}ページ目")
  r = requests.get(url)
  time.sleep(1.5)
  soup = BeautifulSoup(r.text, "html.parser")
  get_a = soup.find_all("a", class_="catalog-search-result-card-header-detail-link color-jd")
  if len(get_a) == 0:
    print(f"全{i}ページ{len(title_links)}科目")
    break

  for j in range(len(get_a)-1):
      try:
          link_ = get_a[j].get("href")
          title_links.append(link_)
      except:
          pass
```

```python
url_data = pd.Series(title_links)
```

```python
strs = ["授業科目名", "学修の目的", "学修内容", "授業改善の工夫", "成績評価方法と基準", "履修上の注意", "その他"]
```

```python
temp = []
for i in range(len(title_links)):
    title_link = title_links[i]
    print(str(i+1)+"科目目",title_link)
    r = requests.get("https://catalog.he.u-tokyo.ac.jp/" + title_link)
    time.sleep(1)
    soup = BeautifulSoup(r.text, "html.parser")
    clses = []
    cls = soup.find_all(class_="catalog-page-detail-lecture-title")
    for j in range(len(cls)):
      cls[j] = cls[j].text.replace('\u3000', '').replace(' ', '')
      cls[j] = cls[j].split('\n')
      clses.append(''.join(cls[j]))
    cls = soup.find_all(class_="catalog-page-detail-lecture-aim")
    for j in range(len(cls)):
      cls[j] = cls[j].text.replace('\u3000', '').replace(' ', '')
      cls[j] = cls[j].split('\n')
      clses.append(''.join(cls[j]))
    cls = soup.find_all(class_="catalog-page-detail-card-body")
    for j in range(len(cls)):
      cls[j] = cls[j].text.replace('\u3000', '').replace(' ', '')
      cls[j] = cls[j].split('\n')
      clses.append(''.join(cls[j]))
    print(clses)
    temp.append(clses)
```

```python
df = pd.DataFrame(temp, columns=strs)
display(df)
```

```python
df.to_csv("tokyo_scraping_result.csv", encoding='utf_8_sig') 
```
